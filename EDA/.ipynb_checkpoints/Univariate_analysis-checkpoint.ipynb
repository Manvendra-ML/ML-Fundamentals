# 1. categorical data valuse
There we can plot data in two ways

## i. Countplot

```python
sns.countplot(x='Survived', data= df2)
```

# ii. piechart


```python
survived_counts = df2['Survived'].value_counts()

plt.pie(
    survived_counts,
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Survival Distribution")
plt.show()
```
